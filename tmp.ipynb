```python
import qrcode
from qrcode.constants import ERROR_CORRECT_M

def make_qr(url: str, outfile: str = "qr.png"):
    qr = qrcode.QRCode(
        version=None,                 # automatic size
        error_correction=ERROR_CORRECT_M,  # 15% restoration
        box_size=10,                  # pixel size of each “box”
        border=4                      # quiet zone (boxes)
    )
    qr.add_data(url)
    qr.make(fit=True)
    img = qr.make_image(fill_color="black", back_color="white")
    img.save(outfile)
    print(f"Saved {outfile}")

make_qr("https://dlai211.github.io/sign-in/index.html", "website_qr.png")
```
